### Webpage url : "https://www.skysports.com/premier-league-table"

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.skysports.com/premier-league-table"
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,"html")
```

```python
table=soup.find("table",class_="standing-table__table")
```

```python
table
```

```python
f_r_e = table.find("tr")
var=f_r_e.find_all("th")[0:10]
```

```python
pieces = [element.text.strip() for element in var]    
```

```python
pieces
```

```python
df = pd.DataFrame(columns=pieces)
df
```

```python
rows = table.find_all("tr")[1:]
rows
```

```python
for row in rows:
    findings = row.find_all("td")[:10]
    elements = [  i.text.strip() for i in findings]
    df=df.append(pd.Series(elements,index=df.columns), ignore_index=True)





```

```python
df
```

```python
df.to_excel(r"Downloads/Premier_league.xlsx",index=False)
```
