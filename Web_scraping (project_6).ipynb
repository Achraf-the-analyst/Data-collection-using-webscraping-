### webpage url : "https://fastestlaps.com/tracks/le-mans-bugatti"

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://fastestlaps.com/tracks/le-mans-bugatti"
```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,"html")
```

```python
table=soup.find("table",class_="table table-striped fl-laptimes-trackpage")
```

```python
elements = table.find_all("th")
elements
```

```python
f_r_e = [element.text.strip()  for element in elements]
f_r_e
```

```python
df=pd.DataFrame(columns=f_r_e)
df
```

```python
rows = table.find_all("tr")[1:]
```

```python
for row in rows :
    pieces = row.find_all("td")
    findings = [piece.text.strip()  for piece in pieces]
    df=df.append(pd.Series(findings,index=df.columns),ignore_index=True)
    
```

```python
df
```

```python
df.to_excel(r"Downloads/Le-Mans-Bugatti.xlsx",index=False)
```
