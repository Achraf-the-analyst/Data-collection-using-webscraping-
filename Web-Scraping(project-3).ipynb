### Dataset:"https://fastestlaps.com/tracks/adm-miachkovo"

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url="https://fastestlaps.com/tracks/adm-miachkovo"
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,"html")
```

```python
table = soup.find_all("table")
table
```

```python
table=soup.find_all("th")
table
```

```python
elements=[]
for element in table:
    elements.append(element.text.strip())
```

```python
df=pd.DataFrame(columns=elements)
```

```python
df
```

```python
rows=soup.find_all("tr")[1:]
rows
```

```python
for row in rows:
    test=row.find_all("td")
    pieces=[piece.text.strip() for piece in test]
    df = df.append(pd.Series(pieces, index=df.columns), ignore_index=True)
```
