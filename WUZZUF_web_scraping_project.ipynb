```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

```python
base_url="https://wuzzuf.net/search/jobs/?a=navb1%7Cspb1%7Cspbg&q=data%20analyst&start="
```

```python
total_pages = 261
```

```python
header = ["job_title","compagny","job_type"]
df=pd.DataFrame(columns=header)
```

```python
for page_number in range(total_pages):
    current_url = base_url + str(page_number)
    page=requests.get(current_url)
    soup = BeautifulSoup(page.text, "html.parser")
    containers = soup.find_all("div",class_="css-1gatmva e1v1l3u10")
    assembles=[]
    for container in containers:
        job = container.find("h2")
        job_title = job.text.strip()
        company = container.find("a",class_="css-17s97q8")
        company_name = company.text.strip()
        job_type = container.find("span",class_="css-1ve4b75 eoyjyou0")
        job_type_vd = job_type.text.strip()
        assembles=[job_title,company_name,job_type_vd]
        df=df.append(pd.Series(assembles,index=df.columns),ignore_index=True)
        time.sleep(0.5)

            
    
    
```
