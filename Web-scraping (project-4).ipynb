### Dataset="https://en.wikipedia.org/wiki/List_of_colors:_A–F"

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/List_of_colors:_A–F"
page=requests.get(url)
```

```python
soup = BeautifulSoup(page.text , "html")
```

```python
table=soup.find_all("table")[0]
```

```python
table
```

```python
elements=table.find_all("th")[0:11]
```

```python
pieces=[]
for element in elements:
    pieces.append(element.text.strip())
    
```

```python
pieces
```

```python
df=pd.DataFrame(columns=pieces)
```

```python
df
```

```python
rows=table.find_all("tr")[1:]
```

```python
for row in rows:
    eliks = row.find_all("th")
    efiks = row.find_all("td")
    pieces_th=[element.text.strip() for element in eliks]
    pieces_td=[element2.text.strip() for element2 in efiks]
    df=df.append(pd.Series(pieces_th + pieces_td ,  index=df.columns) , ignore_index=True)
```
